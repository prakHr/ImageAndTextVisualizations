````python
# flatten the images
n_samples = len(digits.images)
print(n_samples)
data = digits.images.reshape((n_samples, -1))

# Create a classifier: a support vector classifier
# clf = svm.SVC(gamma=0.001)
# clf = KNeighborsClassifier()
# Split data into 50% train and 50% test subsets
X_train, X_test, y_train, y_test = train_test_split(
    data, digits.target, test_size=0.2, shuffle=False
)

clf = LazyClassifier(verbose=0,ignore_warnings=True, custom_metric=None)
models,predictions = clf.fit(X_train, X_test, y_train, y_test)

# print(models)

# # Learn the digits on the train subset
# clf.fit(X_train, y_train)

# # Predict the value of the digit on the test subset
# predicted = clf.predict(X_test)
````

````python
all_possible_models = list(models.index)
scores = list(models.Accuracy)
models_and_scores = list(zip(all_possible_models,scores))
models_and_scores.sort(key = lambda x:x[1])
models_and_scores
````

````python
for m in models_and_scores:
    print(m[0])
````

````python
available_classifier_dict = {
    'DummyClassifier':None,
    'AdaBoostClassifier':ada_boost,
    'ExtraTreeClassifier':extra_trees,
    'GaussianNB':gaussian_nb,
    'DecisionTreeClassifier':decision_tree,
    'QuadraticDiscriminantAnalysis':quadratic_discriminant_analysis,
    'NearestCentroid':None,
    'BernoulliNB':None,
    'RidgeClassifier':None,
    'RidgeClassifierCV':None,
    'PassiveAggressiveClassifier':passive_aggressive,
    'Perceptron':None,
    'BaggingClassifier':None,
    'SGDClassifier':sgd,
    'LinearSVC':svc_linear,
    'XGBClassifier':xgboost_classification,
    'LogisticRegression':None,
    'CalibratedClassifierCV':None,
    'LinearDiscriminantAnalysis':linear_discriminant_analysis,
    'LabelPropagation':None,
    'LabelSpreading':None,
    'NuSVC':None,
    'LGBMClassifier':gradient_boosting,
    'RandomForestClassifier':random_forest,
    'ExtraTreesClassifier':extra_trees,
    'KNeighborsClassifier':knn,
    'SVC':svc,
    
}
````

````python
for i,(name,classifier) in enumerate(list(available_classifier_dict.items())):
    accuracy = models_and_scores[i][1]
    
    if classifier!=None and accuracy<0.90:
        try:
            print(f"Trying to improve the accuracy of worst performing model which in this case is {name} ...")
            estim = HyperoptEstimator(classifier = classifier(name),
                algo=tpe.suggest,
                max_evals=30,
                trial_timeout=120)

            # Search the hyperparameter space based on the data

            estim.fit(X_train, y_train)

            # Show the results

            print( estim.score(X_test, y_test))
            # 1.0

            print(estim.best_model())
            print("*"*100)
        except:
            continue
````
