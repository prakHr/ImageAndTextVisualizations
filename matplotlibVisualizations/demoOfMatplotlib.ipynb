```python
import matplotlib.pyplot as plt
import numpy as np
import math
from math import sqrt
```

```python
# instead of month we can give classes

# instead of sales we can give average feature value according to classes in a variable of array of arrays according to groupby

month = ["Jan","Feb","Mar","Apr","May"]
HYDsales = [50,60,55,75,65]
BNGsales = [55,50,70,60,80]
sales = [HYDsales]
# sales+=21*sales
print(len(sales))
index = np.arange(5)*len(sales)
# width is adjustable between 0 and 1
width = []
for i in range(len(sales)):
    width.append((i+1)/len(sales[i]))
#plt.bar(index,HYDsales,width,color="green",label="Hyd Sales")

for i in range(len(sales)):
    
    plt.bar(index+2*width[i],sales[i],sum(width)/len(width),label = str(i+1))
plt.title("Horizontally Stacked Bar Graphs")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.xticks(index+ sum(width)/len(width), month)
plt.legend(loc="best")
plt.show()
```
